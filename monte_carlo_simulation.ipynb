# Monte Carlo Simulation

## Estimating PI

Throw points randomly in the first quadrant, count how many are inside the circle of radius 1 and multiply by 4 to get an estimate of the true value of pi

$$\frac{\pi}{4}\cdot 4 = \pi$$

```python
def estimate_pi(n, X, Y, F):
  """
  Estimate pi using Monte Carlo simulation
  :param n: number of samples
  :param X: x coordinates of the samples
  :param Y: y coordinates of the samples
  :param F: list of samples inside the circle
  :return: the estimate of pi and the list of samples inside the circle
  """
  for i in range(n):
    if X[i] ** 2 + Y[i] ** 2 <= 1:
      F.append((X[i],Y[i]))
  return 4 * len(F) / n, F
```

```python
def f(x):
  """
  Function of the first quadrant of the circle of radius 1
  :param x: x coordinate
  :return: y coordinate
  """
  return np.sqrt(1 - x ** 2)
```

```python
def pi_simulation(n):
  """
  Simulate pi using Monte Carlo simulation
  :param n: number of samples
  :return: the estimate of pi
  """
  X = np.random.uniform(0, 1, n)
  Y = np.random.uniform(0, 1, n)
  F = []
  pi_prime, F = estimate_pi(n, X, Y, F)
  print("Estimated pi:", pi_prime)
  # Curve
  x = np.linspace(0, 1, n)
  y = f(x)
  plt.plot(x, y, color='r')
  # Samples
  plt.scatter(X, Y)
  # Samples inside
  plt.scatter(*zip(*F))
  # Plot
  plt.xlabel('X')
  plt.ylabel('Y')
  plt.title('Monte Carlo Simulation of PI')
  plt.show()
```

```python
n = 1000
%time pi_simulation(n)
```

# Dice

```python
def dice(n):
  """
  Simulate a dice roll
  :param n: number of sides of the dice
  :return: the result of the roll
  """
  u = np.random.uniform()
  for i in range(n):
    if i/n < u <= (i+1)/n:
      return i+1
```

```python
def dice_simulation(n):
  samples = np.array([dice(6) for i in range(10)])
  # Plot
  plt.hist(samples, align='mid')
  plt.xlabel('Dice Roll')
  plt.ylabel('Frequency')
  plt.title('Histogram of Dice Rolls')
  plt.xticks(range(1, 7))
  plt.show()
```

```python
n = 1000
dice_simulation(n)
```
